# Supervised Learning

The main goal in supervised learning is to learn a model from labeled training data that allows us to make predictions about unseen or future data. Con supervisado se refiere a que "conocemos" para cada ejemplo (X) dado en el conjunto de datos su valor de salid (Y).
Con el modelo construido y testeado se puede a partir de nuevos datos realizar predicciones.

```
Image('02.png', width = 400)
```

A supervised learning task with discrete class labels, such as in the previous email spam filtering example, is also called a classification task. Another subcategory of supervised learning is regression, where the outcome signal is a continuous value.


*   Clasificacin -> El resultado es clasificar un ejemplo X en una etiqueta Y. Ejemplo un mail es SPAM o no es SPAM
*   Regresin -> El resultado es un valor contino. Predecir por ejemplo la temperatura o las ganancias que va a tener un vendedor.
