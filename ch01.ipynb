# Supervised Learning

The main goal in supervised learning is to learn a model from labeled training data that allows us to make predictions about unseen or future data. Con supervisado se refiere a que "conocemos" para cada ejemplo (X) dado en el conjunto de datos su valor de salid (Y).
Con el modelo construido y testeado se puede a partir de nuevos datos realizar predicciones.

```
Image('02.png', width = 400)
```

A supervised learning task with discrete class labels, such as in the previous email spam filtering example, is also called a classification task. Another subcategory of supervised learning is regression, where the outcome signal is a continuous value.


*   Clasificación -> El resultado es clasificar un ejemplo X en una etiqueta Y. Ejemplo un mail es SPAM o no es SPAM
*   Regresión -> El resultado es un valor continúo. Predecir por ejemplo la temperatura o las ganancias que va a tener un vendedor.

##Dimensionality reduction for data compression

```
Image('07.png', width = 500)
```

Deducir la dimensionalidad del problema, a veces tratamos con múltiples dimensiones que lo que hacen es añadir ruido al modelo, lo que se hace entonces es reducir las dimensiones, por ejemplo de 3D a 2D. Es muy común usarlo en el preprocesamiento de características, eliminando características superfluas o para fusionar características que tienen correlación entre ellas.
La reducción de la dimensionalidad consiste en convertir datos de muy alta dimensionalidad en datos de dimensionalidad mucho más baja, de manera que cada una de las dimensiones inferiores transmita mucha más información. 

Otras muchas veces se utiliza simplemente para graficar la información de una manera más clara para poder interpretarla mejor.
